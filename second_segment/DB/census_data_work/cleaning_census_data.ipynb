```python
import pandas as pd
```

```python
# Read in dataset
census_data_df = pd.read_csv("census_estimates.csv")
census_data_df
```

```python
# Get info on df
census_data_df.info()
```

```python
# Drop Null rows
census_data_df.dropna(axis = 0, how="all", inplace=True)
census_data_df
```

```python
# Drop Null Cols
census_data_df.dropna(axis = 1, how="all", inplace=True)
census_data_df
```

```python
# Assert null vals gone
census_data_df.info()
```

```python
# Rename Columns
census_data_df.columns = ['county', "revpop2020", "certpop2021", "popchge", "prcntchge"]
census_data_df
```

```python
# Python code for implementation of upper() 
# Given string and new string 
census_data_df['county'] = census_data_df['county'].str.title()
census_data_df
```

```python
# Drop Unecessary row
census_data_df = census_data_df.drop(index=[0])
```

```python
census_data_df.reset_index()
census_data_df
```

```python
# export to csv
census_data_df.to_csv("cleanCensusData.csv", sep=',', encoding='utf-8', index=False)
```
